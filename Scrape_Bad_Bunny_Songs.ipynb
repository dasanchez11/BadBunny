```python
## numero de canciones
numero = len(driver.find_elements_by_xpath("//div[@class='artista-todas']/div/div[@class='list-container']/ul/li/a"))
numero
```

```python
link_cancion = []
for i in range (0,numero):
    cancion = driver.find_elements_by_xpath("//div[@class='artista-todas']/div/div[@class='list-container']/ul/li/a")[i].get_attribute('href')
    link_cancion.append(cancion)
```

# Sacar la letra de la cancion

```python
driver.get(link_cancion[0])
```

```python
driver.find_elements_by_xpath("//div[@class='cnt-letra p402_premium']/p")[0].text
```

```python
def get_song(driver):
    song = []
    for i in driver.find_elements_by_xpath("//div[@class='cnt-letra p402_premium']/p"):
        estrofa = i.text 
        #+ '\n'
        song.append(estrofa)
    return song
```

```python
link = []
letra = []
for i in link_cancion:
    driver.get(i)
    song_lyric = get_song(driver)
    letra.append(song_lyric)
    link.append(i)
```

```python
import pickle
with open('letras_bad_bu.pickle', 'wb') as l:
    pickle.dump(letra,l)

    
with open('letras_bad_bu_copia.pickle', 'wb') as l:
    pickle.dump(letra,l)
```
